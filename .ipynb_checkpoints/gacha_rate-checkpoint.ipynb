```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.special import comb
from random import choices
```

```python
def bino_dist(rate, roll):
    res = 0
    for i in range(1, roll+1):
        res += comb(roll, i) * rate**i * (1-rate)**(roll-i)
    return [roll, res]
```
